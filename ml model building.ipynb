```python
import numpy as np
import pandas as pd
import tensorflow as tf
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
```

```python
# Load the dataset
loan_data = pd.read_csv('./loan.csv')
```

```python
loan_data
```

```python
loan_data=loan_data.drop(['monthly expenses'],axis=1)
```

```python
loan_data.head()
```

```python
loan_data['Loan_Amount_Term'] = loan_data['Loan_Amount_Term'] / 30
loan_data.head()
```

```python
loan_data['Gender'].fillna(loan_data['Gender'].mode()[0], inplace=True)
loan_data['Married'].fillna(loan_data['Married'].mode()[0], inplace=True)
loan_data['Dependents'].fillna(loan_data['Dependents'].mode()[0], inplace=True)
loan_data['Self_Employed'].fillna(loan_data['Self_Employed'].mode()[0], inplace=True)
loan_data['LoanAmount'].fillna(loan_data['LoanAmount'].mean(), inplace=True)
loan_data['Loan_Amount_Term'].fillna(loan_data['Loan_Amount_Term'].median(), inplace=True)
loan_data['Credit_History'].fillna(loan_data['Credit_History'].mode()[0], inplace=True)

# Encoding categorical variables
loan_data['Dependents'].replace('3+', 3, inplace=True)
```

```python
loan_data.head()
```

```python


loan_data['LoanAmount'] = loan_data['LoanAmount'].round()
loan_data.head()
```

```python
loan_data['ApplicantIncome']*=10
loan_data['LoanAmount']*=4000
```

```python
loan_data.head()
```

```python
import random
# Set the range for random numbers
start = 1.5
end = 2.5
```

```python
n = len(loan_data['ApplicantIncome'])
loan_data['ran_exp'] = loan_data['ApplicantIncome']
for i in range(n):
  ran = random.uniform(start, end)
  loan_data['ran_exp'][i] = loan_data['ApplicantIncome'][i] / ran
  loan_data['ran_exp'] = loan_data['ran_exp'].round()
```

```python
loan_data['remaining'] = loan_data['ApplicantIncome']- loan_data['ran_exp']
loan_data.head()
```

```python
loan_data.head()
```

```python
monthlyRate = 0.01
loan_data['emi'] = loan_data.apply(lambda row:
    (row['LoanAmount'] * monthlyRate * (1 + monthlyRate) ** row['Loan_Amount_Term']) /
    ((1 + monthlyRate) ** row['Loan_Amount_Term'] - 1), axis=1)
loan_data.head()
```

```python
# Create a boolean mask for the condition
mask = loan_data['remaining'] < loan_data['emi']
print(mask.value_counts())
# Set 'Loan_Status' to 'N' where the condition is met
loan_data.loc[mask, 'Loan_Status'] = 'N'
```

```python
loan_data[loan_data['remaining']<loan_data['emi']]['Loan_Status']
```

```python
from sklearn.preprocessing import StandardScaler, LabelEncoder
label_encoders = {}
categorical_columns = ['Gender', 'Married', 'Education', 'Self_Employed']

for column in categorical_columns:
    le = LabelEncoder()
    loan_data[column] = le.fit_transform(loan_data[column])
    label_encoders[column] = le
```

```python
loan_data.head()
```

```python
from sklearn.preprocessing import MinMaxScaler
numerical_features = ['ApplicantIncome', 'LoanAmount','Loan_Amount_Term','ran_exp','remaining','emi']
scaler = MinMaxScaler()
loan_data[numerical_features] = scaler.fit_transform(loan_data[numerical_features])
```

```python
loan_data.head()
```

```python
X = loan_data.drop(columns='Loan_Status')
y = loan_data['Loan_Status'].map({'Y': 1, 'N': 0})
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.20, random_state=42)
```

```python
X_train.head()
```

```python
X_train_tensor = tf.convert_to_tensor(X_train, dtype=tf.float32)
Y_train_tensor = tf.convert_to_tensor(y_train, dtype=tf.float32)

X_test_tensor = tf.convert_to_tensor(X_test, dtype=tf.float32)
y_test_tensor = tf.convert_to_tensor(y_test, dtype=tf.float32)
```

```python
model1 = tf.keras.models.Sequential([
    tf.keras.layers.Dense(64, activation='relu', input_shape=(X_train.shape[1],)),
    tf.keras.layers.Dense(32, activation='relu'),
    tf.keras.layers.Dense(16, activation='relu'),
    tf.keras.layers.Dense(1, activation='sigmoid')
])

model1.compile(optimizer='adam', loss='binary_crossentropy', metrics=['accuracy'])
```

```python
history = model1.fit(X_train_tensor, Y_train_tensor, epochs=50, batch_size=32, validation_split=0.2)

# Evaluating the model
loss, accuracy = model1.evaluate(X_test_tensor, y_test_tensor)
print('Test accuracy:', accuracy)
print('Test loss:', loss)
```

```python
model2 = tf.keras.models.Sequential([
    tf.keras.layers.Dense(128, activation='relu', input_shape=(X_train.shape[1],)),
    tf.keras.layers.Dropout(0.3),
    tf.keras.layers.Dense(64, activation='relu', kernel_regularizer=tf.keras.regularizers.l2(0.01)),
    tf.keras.layers.Dropout(0.3),
    tf.keras.layers.Dense(32, activation='relu', kernel_regularizer=tf.keras.regularizers.l2(0.01)),
    tf.keras.layers.Dense(1, activation='sigmoid')
])

model2.compile(optimizer='adam', loss='binary_crossentropy', metrics=['accuracy'])

# Using early stopping to avoid overfitting
early_stopping = tf.keras.callbacks.EarlyStopping(monitor='val_loss', patience=5, restore_best_weights=True)

# Training the model
history = model2.fit(X_train, y_train, epochs=100, batch_size=32, validation_split=0.2, callbacks=[early_stopping])
```

```python
loss, accuracy = model2.evaluate(X_test_tensor, y_test_tensor)
print('Test accuracy:', accuracy)
print('Test loss:', loss)
```
